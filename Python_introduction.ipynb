### 2.2 Basic data structures

#### **List**
Lists are exactly what they sound like: they are containers that house elements in a given sequence/order.

````python
number_list = [1, 2, 3, 4, 5]
string_list = ['a', 'p', 'p', 'l', 'e']
mixed_list = ['orange', 22, 'f', 67.2]
````

How could we access the objects that we store in lists?

````python
# indexing
print(number_list[0]) # return the first item of the list, note python indexing starts from 0
print(number_list[-1]) # return the last item of the list
print(number_list[1:3]) # return 2nd to 3rd item of the list, note number_list[3] would not be printed
print(number_list[:3]) # return 1st to 3rd item of the list
# concatenation
print(number_list + string_list) # return a list containing all elements in string_list and number_list
````

Here are some useful functions for list operations:


> Get the length of a list: `len()`.  
Get the maximum value in a list: `max()`.  
Check if a certain element or value is in a list: `in`.  
Sort the list: `sorted()`.  
Count the instances of an element in the list: `.count()`.  
Coerce strings to a list: `list()`

Further documentation: https://www.programiz.com/python-programming/methods/list

````python
a = [1,54, 8, 0, 0,  3, 99]
print(len(a))
print(max(a))
print(9 in a)
print(sorted(a))
print(a.count(0))
b = 'aaaaa'
print(list(b))
````

#### **Dictionary**
Dictionaries hold key:value pairs - in some languages these are called "hash tables".

Dictionaries can store a mixture of different "types": all keys and values can be a mixture of types. Note here keys could not be unhashable types (so could not be data structures like list, set, etc.)

````python
student_info_dict = {2: 120120,
                     "first_name": ["Jane","Marry"],
                     "last_name": "Dough"}
print(student_info_dict)
````

````python
# indexing with keys (could not index with values)
print(student_info_dict[2])
print(student_info_dict['first_name'][0])
# change values
student_info_dict['first_name'][0] = 1
print(student_info_dict['first_name'][0])
# iterating through dictionaries
for key, value in student_info_dict.items():
    print(f"{key}: {value}")
for key in student_info_dict.keys():
    print(f"{key}: {student_info_dict[key]}")
````

### 3.3 Indexing

Now we need to understand the indexing of a dataframe (the index appears as the first column of the dataframe, starting from 0 by default)

Columns of DataFrames are much more intuitive to index than columns of arrays. You can select the values of a single column by indexing with the column label. This is called a "series".

```
dataframe[column_label]
```

<img src ='https://github.com/ccbskillssem/pythonbootcamp/raw/main/day_4/ColumnIndex.png'>

````python
# select the column with dogs colors
heart_disease_df["Age"]
````

DataFrame rows are indexed similarly to array rows, with a minor syntax difference. With DataFrames, we use `.loc[]` instead of `[]` to access rows.

```
dataframe.loc[row_label]
```

Just as with columns, rows of DataFrames are also stored in Series structures: the only difference is that the column labels have been transposed to row labels, as a Series only has one Index for row labels.

<img src='https://raw.githubusercontent.com/ccbskillssem/pythonbootcamp/main/day_4/RowIndex.png'>

It can help to think of a DataFrame as a Lego-like structure composed of Series. When we select parts of the DataFrame, we're snapping off some of the Legos from the main structure.

````python
# select the row with index 2
heart_disease_df.loc[2]
````

#### (a) Scatter plot

Let's try to see if there is any correlation between age and level of serum cholesterol for the female population

````python
from plotnine import ggplot, aes, geom_point, theme_bw, theme
from matplotlib import pyplot as plt

(ggplot(heart_disease_df[heart_disease_df.Sex != 1], aes(x='Serum_Cholesterol', y='Age', color='Diagnosis_Heart_Disease')) +
    geom_point(size=1.5, alpha=0.5) +
    theme_bw() +
    theme(figure_size=(6, 3)))
````

````python
heart_disease_df['Diagnosis_Heart_Disease'] = heart_disease_df['Diagnosis_Heart_Disease'].astype('category')
````

Now rerun the above cell again, you will find that the color becomes discrete

Facet plots are figures made up of multiple subplots which share the same set of axes. This is useful when you want to make side-by-side comparisons.

````python
from plotnine import ggplot, aes, geom_point, facet_grid, facet_wrap, theme_bw, theme, element_text

(ggplot(heart_disease_df, aes(x='Age', y='Serum_Cholesterol')) +
      geom_point(size=1, alpha=0.5) +
      facet_wrap('~Diagnosis_Heart_Disease', scales='free_y') +
      theme_bw() +
      theme(axis_text_x=element_text(rotation=90, hjust=1),  # hjust should be within [0, 1]
            figure_size=(6, 3)))  # Set the figure size here properly
````

#### (b) Histogram
Histograms are good for showing general distributional features of dataset variables. You can see roughly where the peaks of the distribution are, whether the distribution is skewed or symmetric, and if there are any outliers.

Let's explore the age distribution of our dataset

````python
from plotnine import ggplot, aes, geom_histogram, theme_bw, xlab, ylab

# Create the plot
plt = (ggplot(heart_disease_df, aes(x='Age')) +
       geom_histogram(bins=30) +
       theme_bw() +
       xlab('Age') +   # Change to correct label, if the column is 'Age'
       ylab('Count'))
# Display the plot
display(plt)

# ggsave(filename='/content/gdrive/My Drive/histogram.png', plot=plt)
````

#### (c) Box plot

Box plots are used to visualize the distribution of numeric data values, especially when comparing multiple groups

````python
# Create the 'Age_group' column based on conditions
heart_disease_df['Age_group'] = pd.cut(heart_disease_df['Age'],
                                       bins=[-float('inf'), 35, 65, float('inf')],
                                       labels=['<35', '35-65', '>65'])
````

````python
from plotnine import ggplot, aes, geom_boxplot, geom_jitter, theme_bw, xlab, ylab, theme, facet_wrap

# Create the plot
plt = (ggplot(heart_disease_df, aes(x='Age_group', y='Serum_Cholesterol')) +
       geom_boxplot(outlier_size=0, outlier_color='white') +  # Use outlier_size=0 to hide outliers
       geom_jitter(size=0.5, width=0.2, color='red') +
       ylab('Serum Cholesterol') +
       xlab('Age Group') +
       theme_bw() +
       theme(figure_size=(8, 4)) +
       facet_wrap('~ Sex', scales='free_x'))

# Display the plot
display(plt)
````

#### (d) Heatmap

````python
from plotnine import ggplot, aes, geom_tile, geom_text, theme_bw, xlab, ylab, theme
from matplotlib import pyplot as plt

heart_disease_df['Sex'] = heart_disease_df['Sex'].astype('category') # make column 'Sex' categorical datatype

plt = (ggplot(heart_disease_df, aes(x='Sex', y='Age', fill='Serum_Cholesterol'))+
    geom_tile()+
    geom_text(aes(label='Serum_Cholesterol'), size=6, color='white')+
    xlab('Sex (female: 0, male:1)')+
    ylab('Age')+
    theme_bw()+
    theme(figure_size=(4, 10)))
display(plt)
#ggsave(plt, '/content/gdrive/My Drive/WBE_UCB/Papers/testing_laboratory_blueprint/quality_score_summary_heatmap.png')
````
